```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

#load dataset
data = pd.read_csv('Students_Grading_Dataset.csv')

#input and output
X = data[['Study_Hours_per_Week']]
y = data['Final_Score']

#train model
model = LinearRegression()
model.fit(X, y)
predicted_scores = model.predict(X)

#valid regression metrics
mae = mean_absolute_error(y, predicted_scores)
mse = mean_squared_error(y, predicted_scores)
rmse = np.sqrt(mse)
r2 = r2_score(y, predicted_scores)

print('MAE: ', round(mae, 2))
print('MSE: ', round(mse, 2))
print('RMSE: ', round(rmse, 2))
print('R2 Score: ', round(r2, 4))

plt.figure(figsize=(10, 6))
plt.hist(data['Final_Score'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Fianl Exam Scores')
plt.xlabel('Fianl Exam Scores')
plt.ylabel('Number of Students')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='orange', label='Actual Scores')
plt.plot(X, predicted_scores, color= 'red', label='Predicted Scores')
plt.title('Model Prediction vs Actual Scores')
plt.xlabel('Study Hours Per Week')
plt.ylabel('Final Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```
